```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

def predict(model, img_path, class_names):
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(256, 256))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)

    predictions = model.predict(img_array)
    predicted_class = class_names[np.argmax(predictions[0])]
    confidence = round(100 * np.max(predictions[0]), 2)

    return predicted_class, confidence


model = tf.keras.models.load_model(r"C:\Users\frede\agricultura\model\20")


class_names = ['miner', 'nodisease', 'phoma', 'rust']


image_dir1 = r"C:\Users\frede\agricultura\test\miner"
image_dir2= r"C:\Users\frede\agricultura\test\nodisease"
image_dir3=r"C:\Users\frede\agricultura\test\phoma"
image_dir4=r"C:\Users\frede\agricultura\test\rust"

def imprimir(model, image_dir, class_names):
    for filename in os.listdir(image_dir):
        if filename.endswith('.jpg'):
      
            image_path = os.path.join(image_dir, filename)

     
            predicted_class, confidence = predict(model, image_path, class_names)


            plt.figure(figsize=(6, 6))
            img = plt.imread(image_path)
            plt.imshow(img)
            plt.title(f"Predicted Class: {predicted_class}\nConfidence: {confidence}%")
            plt.axis('off')
            plt.show()
 
def taxa_acertos(model, image_dir, class_names, resposta):
    acertos=0
    erro=0
    resposta
    for filename in os.listdir(image_dir):
        
        image_path = os.path.join(image_dir, filename)
        
        predicted_class, confidence = predict(model, image_path, class_names)
        
        if predicted_class==resposta:
            acertos+=1
        else:
            erro+=1
    print(acertos/(acertos+erro))
```

```python
#miner
taxa_acertos(model, image_dir1, class_names,resposta='miner')
```

```python
#nodisease
taxa_acertos(model, image_dir2, class_names,resposta='nodisease')
```

```python
#phoma
taxa_acertos(model, image_dir3, class_names,resposta='phoma')
```

```python
#rust
taxa_acertos(model, image_dir4, class_names,resposta='rust')
```
